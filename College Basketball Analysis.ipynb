# College Basketball Analysis

For my project, I decided to use a men’s college basketball database that I found on Kaggle.com. The dataset is https://www.kaggle.com/andrewsundberg/college-basketball-dataset. The dataset has a usability score of 10 and has very clean data. This dataset contains information pertaining to the 2013-2021 seasons for men’s Division I. It includes 25 variables which are RK: the team ranking at the end of the regular season, TEAM: the team name of the Division 1 school, CONF: the conference each team plays in, G: the number of games played in that season, W: the number of games won that season, ADJOE: the adjusted offensive efficiency (points scored per 100 possessions), ADJDE: the adjusted defensive efficiency (points allowed per 100 possessions), BARTHAG: the power rating (chance at beating an average D1 team), EFG_O: effective field goal percentage shot, EFG_D: effective field goal percentage against, TOR: turnover rate, TORD: steal rate, ORB: offensive rebound rate, DRB: offensive rebound rate allowed, FTR: free throw rate (how often a team shoots free throws), FTRD: free throw rate allowed, 2P_O: two point shooting percentage, 2P_D: two point shooting percentage allowed, 3P_O: three point shooting percentage, 3P_D: three point shooting percentage allowed, ADJ_T: adjusted tempo (tempo per 40 minutes a team would have against the team), WAB: wins above bubble (cut off of making tournament), POSTSEASON: postseason (round where team was eliminated from the tournament or where their season ended), SEED: their seed ranking in the March Madness Tournament, and YEAR: the year the data is displaying for that team.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a dataframe of the cbb csv (2013-2019 years)
basketball_stats_df = pd.read_csv(data + '/cbb.csv')
basketball_stats_df
```

I then looked through the variable datatypes to see what they were defined as. 

```python
basketball_stats_df.dtypes
```

Following this, I made sure to only look at the tournament winners (7 in total). These teams were Louisville (2013), Connecticut (2014), Duke (2015), Villanova (2016 and 2018), North Carolina (2017), and Virginia (2019). I also altered the number of visible columns to 24. This allowed me to view all the columns at once. 

As of now, I have only created charts for the tournament winners, but I am considering altering this and simply highlighting the winners on the charts. I Could also make separate charts that compare runners-up to champions, etc. 

```python
# Prints only the winners of the tournament
tournament_win = basketball_stats_df['POSTSEASON'] == 'Champions'

tournament_winners_df = basketball_stats_df[tournament_win]
# Sets the column max to 24 so all columns are visible
pd.set_option("display.max_columns", 24)
tournament_winners_df
```

The first charts I created are comparing the seven champions adjusted offensive and defensive efficiencies vs their total wins for the season they won. I plan to clean up the charts by editing the decimals to whole numbers and highlighting which team is which data point. I am also considering changing this to one chart that highlights both efficiencies.

The data from these charts shows that the more wins a team has, either their offensive or defensive efficiencies were high, if not both. 

```python
# Create scatter plots for offensive and defensive efficiencies vs wins
fig = plt.figure(figsize = (12, 6))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
    
ax1.scatter(tournament_winners_df['W'], tournament_winners_df['ADJOE'], color = 'blue')
ax1.set_title('Adjusted Offensive Efficiency vs. Wins')
ax1.set_xlabel('Wins')
ax1.set_ylabel('Offensive Efficiency')

ax2.scatter(tournament_winners_df['W'], tournament_winners_df['ADJDE'], color = 'red')
ax2.set_xlabel('Wins')
ax2.set_ylabel('Defensive Efficiency')
ax2.set_title('Adjusted Defensive Efficiency vs. Wins')

plt.show()
plt.close()
```

```python
# Creates a scatter plot for offensive efficiency by wins for all teams
fig, ax = plt.subplots(figsize = (14,10))
basketball_stats_df.plot.scatter(x = "W",
                                 y = "ADJOE", 
                                 s = 30, 
                                 ax = ax)

plt.title("Offensive Efficiency vs Wins", fontsize = 26)
ax.set_xlabel("Wins", fontsize = 20)
ax.set_ylabel("Offensive Efficiency", fontsize = 20)

plt.show()
plt.close()
```

```python
# Looks at top 16 teams in the tournament for each year
top_sixteen = basketball_stats_df[(basketball_stats_df["POSTSEASON"] == "Champions") |
                                  (basketball_stats_df["POSTSEASON"] == "2ND") |
                                  (basketball_stats_df["POSTSEASON"] == "F4") |
                                  (basketball_stats_df["POSTSEASON"] == "E8") |
                                  (basketball_stats_df["POSTSEASON"] == "S16")]
top_sixteen
```

```python
# Creates a scatter plot for offensive efficiency by wins for top 16 teams in the tournament each year
fig, ax = plt.subplots(figsize = (14,10))
top_sixteen.plot.scatter(x = "W",
                         y = "ADJOE", 
                         s = 30, 
                         ax = ax)

plt.title("Offensive Efficiency vs Wins", fontsize = 26)
ax.set_xlabel("Wins", fontsize = 20)
ax.set_ylabel("Offensive Efficiency", fontsize = 20)

plt.show()
plt.close()
```

```python
# Shows all teams with a positive wins above bubble
positive_WAB = basketball_stats_df.loc[basketball_stats_df["WAB"] > 0]
positive_WAB
```

```python
# Creates a scatter plot for WAB and wins for top 16 teams in the tournament
fig, ax = plt.subplots(figsize = (14,10))
top_sixteen.plot.scatter(x = "W",
                         y = "WAB", 
                         s = 30, 
                         ax = ax)

plt.title("WAB and Wins", fontsize = 26)
ax.set_xlabel("Wins", fontsize = 20)
ax.set_ylabel("Wins Above Bubble", fontsize = 20)

plt.show()
plt.close()
```

```python
# Creates a scatter plot for WAB and wins
fig, ax = plt.subplots(figsize = (14,10))
positive_WAB.plot.scatter(x = "W",
                         y = "WAB", 
                         s = 30, 
                         ax = ax)

plt.title("WAB and Wins", fontsize = 26)
ax.set_xlabel("Wins", fontsize = 20)
ax.set_ylabel("Wins Above Bubble", fontsize = 20)

plt.show()
plt.close()
```

My second group of charts displays the teams’ effective field goal percentage for and against on average vs wins. I am considering altering these in a similar way as the previous two charts. 

These charts show that teams had considerably higher effective field goal percentages for themselves compared to against them for their opponents. This was expected because as a team’s offensive efficiency goes up, so should their field goal percentage. This is the same with the defensive side. 

```python
# Create scatter plots for effective field goal percentage for and against vs wins
fig = plt.figure(figsize = (12, 6))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
    
ax1.scatter(tournament_winners_df['W'], tournament_winners_df['EFG_O'], color = 'blue')
ax1.set_title('Effective Field Goal Percentage')
ax1.set_xlabel('Wins')
ax1.set_ylabel('Effective FG %')

ax2.scatter(tournament_winners_df['W'], tournament_winners_df['EFG_D'], color = 'red')
ax2.set_title('Effective Field Goal Percentage Against')
ax2.set_xlabel('Wins')
ax2.set_ylabel('Effective FG % Against')

plt.show()
plt.close()
```

Next, I created bar charts that compared each years winning team with their turnover rate and their steal rate. I am considering altering these charts to a single chart displaying each teams’ turnovers and steals next to each other. This would probably make it easier to read the data. This can show whether teams had a positive ratio. 

These charts show that for each team, they averaged more steals per game than turnovers per game. This can be used to point out that winning teams tend to cause more turnovers than they have themselves which will increase their chances of winning a game. 

```python
# Bar chart displaying teams turnover rate
plt.bar(tournament_winners_df['YEAR'], tournament_winners_df['TOR'])
plt.title('Turnover Rate by Year')
plt.xlabel('Year')
plt.ylabel('Turnover Rate')

plt.show()
plt.close()
```

```python
# Bar chart displaying team steal rate
plt.bar(tournament_winners_df['YEAR'], tournament_winners_df['TORD'])
plt.title('Steal Rate by Year')
plt.xlabel('Year')
plt.ylabel('Steal Rate')

plt.show()
plt.close()
```

The following charts display each team’s two point shooting percentage for and against per game on average. If a team can shoot a higher percentage, they are more efficient at scoring, meaning they can score more points per possession. These charts drill down to more specifics than the overall effective field goal percentage and offensive and defensive efficiencies.The championship winning teams on average had a higher two point shooting percentage than their opponents. 

Once again, I plan to add to these charts by editing the axes and adding a way to see which team is which. 

```python
# Create scatter plots for 2 point shooting percentage for and against
fig = plt.figure(figsize = (12, 6))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
    
ax1.scatter(tournament_winners_df['G'], tournament_winners_df['2P_O'], color = 'blue')
ax1.set_title('Two Point Shooting Percentage')
ax1.set_xlabel('Games Played')
ax1.set_ylabel('2 Pt Shooting %')

ax2.scatter(tournament_winners_df['G'], tournament_winners_df['2P_D'], color = 'red')
ax2.set_title('Two Point Shooting Percentage Against')
ax2.set_xlabel('Games Played')
ax2.set_ylabel('2 Pt Shooting % Against')

plt.show()
plt.close()
```

The following charts display each team’s three point shooting percentage for and against per game on average. If a team can shoot a higher percentage, they are more efficient at scoring, meaning they can score more points per possession. This is more important for shooting threes compared to two pointers because you get one more point per shot. As shown previously with two point shooting, on average, the March Madness Tournament winning teams shot a higher three point percentage than their opponents. 

```python
# Create scatter plots for 3 point shooting percentage for and against
fig = plt.figure(figsize = (12, 6))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
    
ax1.scatter(tournament_winners_df['G'], tournament_winners_df['3P_O'], color = 'blue')
ax1.set_title('Three Point Shooting Percentage')
ax1.set_xlabel('Games Played')
ax1.set_ylabel('3 Pt Shooting %')

ax2.scatter(tournament_winners_df['G'], tournament_winners_df['3P_D'], color = 'red')
ax2.set_title('Three Point Shooting Percentage Against')
ax2.set_xlabel('Games Played')
ax2.set_ylabel('3 Pt Shooting % Against')

plt.show()
plt.close()
```
