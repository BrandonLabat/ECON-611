```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dataframe of the cbb csv (2013-2019 years)
basketball_stats_df = pd.read_csv(data + '/cbb.csv')
basketball_stats_df
```

```python
basketball_stats_df.dtypes
```

The POSTSEASON column made it difficult to plot because of how it was structured, so I switched replaced each value with a number. 1 = Champion, 2 = 2nd, 4 = final four appearance, 8 = elite eight appearance, 16 = sweet 16, 32 = round of 32, 64 = round of 64, and 68 = round of 68. 

```python
# Changes POSTSEASON values to make it easier to plot
basketball_stats_df["POSTSEASON"].replace({"Champions": "1", "2ND": "2", "F4": "4", "E8": "8", "S16": "16", "R32": "32", "R64": "64", "R68": "68"}, inplace = True)
basketball_stats_df
```

Also, I made these numbers for POSTSEASON categories to make ordering them on graphs simpler. When this was not done, the graphs would plot the season outcome in alphabetical order. 

```python
# Allows POSTSEASON to be plotted in order
basketball_stats_df.rename(columns={21:'POSTSEASON'}, inplace = True)
basketball_stats_df['POSTSEASON'] = pd.Categorical(basketball_stats_df['POSTSEASON'], 
                                                   categories = ['1', '2', '4', '8', '16', '32', '64', '68'], ordered = True)
basketball_stats_df = basketball_stats_df.sort_values("POSTSEASON", ascending = False)
```

Following this, I made sure to only look at the tournament winners (7 in total). These teams were Louisville (2013), Connecticut (2014), Duke (2015), Villanova (2016 and 2018), North Carolina (2017), and Virginia (2019). I also altered the number of visible columns to 24. ### March Madness Teams WAB vs Season Outcome

Once I finished looking into the offensive and defensive efficiencies, I created a scatter plot comparing wins above the bubble and the season outcome of each team (Where each team was eliminated from the tournament). Wins above the bubble is a statistical variable that determines how often a team would be expected to win each game. This is based on the team opponent and the location of the game. For example, if a team has a record of 12-3 and a bubble team would be expected to be 9-6 based on the opponents and locations of the games, the team with a 12-3 record has a WAB rating of 3. This chart shows that the championship teams are near the top in WAB for teams in the tournament (excluding the Connecticut Huskies in 2014). This proves that the championship teams can win a large quantity of games that tournament teams are supposed to win. These wins aid a team in climbing the rankings and gaining confidence. 

```python
# Creates a scatter plot for WAB vs season outcome for all teams in the tournament
fig, ax = plt.subplots(figsize = (12,8))
top_sixtyeight.plot.scatter(x = "POSTSEASON",
                         y = "WAB", 
                         s = 30, 
                         ax = ax)

plt.title("WAB vs Season Outcome", fontsize = 26)
ax.set_xlabel("Season Outcome", fontsize = 20)
ax.set_ylabel("Wins Above Bubble", fontsize = 20)
# Sets the axis labels
plt.ylim(-3, 13)
# Labels each team
for i in range(tournament_winners_df.shape[0]):
    plt.annotate(tournament_winners_df.TEAM.tolist()[i],
                 (tournament_winners_df.POSTSEASON.tolist()[i],
                 tournament_winners_df.WAB.tolist()[i]))
print(tournament_winners_df.shape[0])
plt.tight_layout() 

plt.show()
plt.close()
```

### Effective FG % Tournament Teams

The charts after WAB are displaying the teams in the tournament’s effective field goal percentages and their opponents effective field goal percentage for the total amount of games they played throughout the season. Effective field goal percentage accounts for three point shots being worth more than two point shots. The formula is (FGM + 0.5 * 3PM) / FGA. Every championship team’s effective field goal percentage is higher than their opponents. This is expected because these teams all had relatively high offensive efficiencies and low defensive efficiencies. As a team’s offensive efficiency goes up, so should their field goal percentage. This is the same for the defensive side. 

```python
# Creates a scatter plot for effective field goal % by games played for tournament teams
fig, ax = plt.subplots(figsize = (12, 8))
top_sixtyeight.plot.scatter(x = "G",
                         y = "EFG_O", 
                         s = 30,
                         ax = ax)

plt.title("Effective Field Goal Percentage", fontsize = 26)
ax.set_xlabel("Games", fontsize = 20)
ax.set_ylabel("EFG %", fontsize = 20)
# Highlights the tournament winners in red
plt.scatter(tournament_winners_df['G'], tournament_winners_df['EFG_O'], color = "red")
# Labels each team
for i in range(tournament_winners_df.shape[0]):
    plt.annotate(tournament_winners_df.TEAM.tolist()[i],
                 (tournament_winners_df.G.tolist()[i],
                 tournament_winners_df.EFG_O.tolist()[i]))
print(tournament_winners_df.shape[0])
plt.tight_layout() 

plt.show()
plt.close()
```

### Offensive Rebound Rate

The last charts I created are the tournament team’s offensive rebound rate for and against. Grabbing rebounds is important to give your team more opportunities to score. The higher percentage of times your team can grab a rebound from your team’s missed shot, the more opportunities there are to score. This is also true on the defensive side. The more often your team can prevent the opponent from grabbing an offensive rebound, the less likely your opponent is to score because they do not receive as many opportunities. The offensive rebound rate is not very correlated to whether a team wins the championship or not. There are a few teams that have very high percentages, but there are several that are not nearly as successful at this. 

```python
# Creates a scatter plot for offensive rebound rate
fig, ax = plt.subplots(figsize = (12,8))
top_sixtyeight.plot.scatter(x = "YEAR",
                                 y = "ORB", 
                                 s = 30, 
                                 ax = ax)

plt.title("Offensive Rebound Rate", fontsize = 26)
ax.set_xlabel("Year", fontsize = 20)
ax.set_ylabel("ORB", fontsize = 20)
# Highlights the tournament winners in red
plt.scatter(tournament_winners_df['YEAR'], tournament_winners_df['ORB'], color = "red")
# Labels each team
for i in range(tournament_winners_df.shape[0]):
    plt.annotate(tournament_winners_df.TEAM.tolist()[i],
                 (tournament_winners_df.YEAR.tolist()[i],
                 tournament_winners_df.ORB.tolist()[i]))
print(tournament_winners_df.shape[0])
plt.tight_layout() 

plt.show()
plt.close()
```
