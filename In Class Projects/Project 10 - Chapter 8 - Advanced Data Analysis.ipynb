## Creating Indicator Variables

```python
def create_indicator_variable(data,
                              indicator_name,
                              index_name,
                              target_index_list):
    # Prepare column with name of the indicator variable
    data[indicator_name] = 0
    # For each index's name that matches an entry in target_index_list,
    # value of 1 will be recorded
    # for index in target_index_list:
        # .loc checks row entries to see if entity index matches
        # the entity from the target index list
        # Sets values of indicator_name (column) to 1
        # for all such rows 
    data.loc[target_index_list, [indicator_name]] = int(1)

data = pd.read_csv("fraserDataWithRGDPPC.csv",
                   index_col = ["ISO_Code", "Year"],
                   parse_dates = True)

# Select ISO_Code from names of double index
index_name = data.index.names[0]
indicator_name = "North America"
countries_in_north_america = [
    "BHS", "BLZ", "CAN", "CRI", "DOM", "SLV",
    "GTM", "HTI", "HND", "JAM", "MEX", "NIC", "PAN",
    "TTO", "USA"]
create_indicator_variable(data = data,
                          indicator_name = indicator_name,
                          index_name = index_name,
                          target_index_list = countries_in_north_america)
```

## Creating a Quantile Ranking

```python
# Number of divisions
n = 5
# Creates a column identifying n-tile rank
quantile_var = "RGDP Per Capita"
quantile_name = quantile_var + " " + str(n) + "-tile"
data[quantile_name] = np.nan
data
```

```python
def create_quantile(n, 
                    data, 
                    year, 
                    quantile_var, 
                    quantile_name):
    # Identifies countries for a given year
    year_index = data.index.get_level_values("Year") == year
    # Selects the subset of the dataframe by year and variable
    year_data = data[year_index][quantile_var]
    # Creates a dictionary that holds the quantile values
    quantile_values_dict = {i:year_data.quantile(i/n) for i in range(1, n + 1)}
    
    # Cycles through each country for the year
    for index in data[year_index].index:
        # Identifies the value of the variable
        val = data.loc[index][quantile_var]
        # Compares the value to the values that divide each quantile
        for i in range(1, n + 1):
            # Saves the quantile as i if the value is < highest in the quantile
            if val <= quantile_values_dict[i]:
                data.loc[index,[quantile_name]] = int((n + 1) - i)
                break
            # Otherwise checks the highest value of the next quantile
            else:
                continue
                
years = data.groupby("Year").mean().index
for year in years:
    create_quantile(n, 
                    data, 
                    year, 
                    quantile_var, 
                    quantile_name)
```
