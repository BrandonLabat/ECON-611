# Forming an array with a list of different types of values

```python
# Has one value that is not an int
int_float_list = [1,2,3,4.,5]
print("int_float_list")
for val in int_float_list:
    print(val, type(val))

# All values are changed to float 64 type
mystery_array = np.array(int_float_list)  
print() 
print("mystery_array")
for val in mystery_array:
    print(val, type(val))

# Has all values as float 64 type
float_list = [1.,2.,3.,4.,5.]
float_array = np.array(float_list)
print()
print("float_array")
for val in float_array:
    print(val, type(val))
```

In last case (float_array), we check to see if the transfomration of a list of all floats is the same as the transformation of a mixed list of floats and ints (int_float_list).

# Monte Carlo Simulation

First, must build a list of points drawn from a standard normal distribution

```python
random_list = []
# Sets the range so we can change the value easier later on
sim_length = 1000
# Define variables here so it is easier later on
mean = 0
sigma = 0.5

for i in range(sim_length):
    random_list.append(random.normalvariate(mean, sigma))
plt.plot(random_list, ls = "", marker = ".", markersize = 2)
plt.show()
plt.close()
```

# Single Simulation

Take cumulative sum of the list at each index to create a randomly generated line

```python
simulation = []

for i in range(len(random_list)):
    val = random_list[i]
    if i == 0:
        simulation.append(val)
    else:
        simulation.append(simulation[i - 1] + val)
plt.plot(simulation)
plt.show()
plt.close()
```

# Building 1000 Single Simulations

Records each result in a dictionary with each period of the simulation being represented in the second layer of the dictionary of dictionaries. 

```python
monte_carlo_sim_dict = {}
num_sims = 2000
periods = 1000

for i in range(num_sims):
    monte_carlo_sim_dict[i] = {}
    for period in range(periods):
        # current sim = curr_sim: So we don't have to type monte.... out everytime
        # Shorthand simulation
        curr_sim = monte_carlo_sim_dict[i]
        if period == 0:
            curr_sim[period] = random.normalvariate(mean, sigma)
        else:
            curr_sim[period] = curr_sim[period - 1] + random.normalvariate(mean, sigma)
```

```python
# Transforms the dataframe
monte_carlo_sim_df = pd.DataFrame(monte_carlo_sim_dict)
monte_carlo_sim_df
```
