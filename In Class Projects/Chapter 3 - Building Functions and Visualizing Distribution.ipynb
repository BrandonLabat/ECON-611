## Building Functions

```python
def total(lst_obj):
    total = 0
    n = len(lst_obj)
    for i in range(n):
        total += lst_obj[i]
        # Use below code to print total after each i is added
        #print(i, total)
    return total

def mean(lst_obj):
    n = len(lst_obj)
    mean_ = total(lst_obj) / n
    return mean_

def median(lst_obj):
    n = len(lst_obj)
    lst_obj = sorted(lst_obj)
    if n % 2 != 0:
        # Uses if list length is odd
        middle_index = int((n - 1) / 2)
        median_ = lst_obj[middle_index]
    else:
        upper_middle_index = int(n / 2)
        lower_middle_index = upper_middle_index - 1
        # Passes slice with two middle values to mean()
        median_ = mean(lst_object[lower_middle_index : upper_middle_index + 1])
    return median_

def mode(lst_obj):
    # To record values that appear the most
    max_count = 0
    counter_dict = {}
    for value in lst_obj:
        counter_dict[value] = 0
    for value in lst_obj:
        counter_dict[value] += 1
    count_list = list(counter_dict.values())
    max_count = max(count_list)
    mode_ = [key for key in counter_dict if counter_dict[key] == max_count]
    return mode_

def variance(lst_obj, sample = False):
    # Saves the mean value of list
    list_mean = mean(lst_obj)
    # n is used to calculate the average of sum squared differences
    n = len(lst_obj)
    # Value that can add squared differences to
    sum_sq_diff = 0
    for val in lst_obj:
        # Adds each squared difference to sum_sq_diff
        sum_sq_diff += (val - list_mean) ** 2
    if sample == False:
        # Normalizes the result by dividing by n
        variance_ = sum_sq_diff / n
    else:
        # For samples, normalizes by dividing by (n - 1)
        variance_ = sum_sq_diff / (n - 1)
    return variance_

def SD(lst_obj, sample = False):
    # Standard deviation = square root of variance
    SD_ = variance(lst_obj, sample) ** (1/2)
    
    return SD_

def covariance(lst_obj1, lst_obj2, sample = False):
    # Determines the mean of each list
    mean1 = mean(lst_obj1)
    mean2 = mean(lst_obj2)
    # Creates a variable = 0 that is used to sum the values generated in the for loop
    cov = 0
    n1 = len(lst_obj1)
    n2 = len(lst_obj2)
    # Checks to see if lengths are equal
    if n1 == n2:
        n = n1
        # Sum the product of the differences
        for i in range(n1):
            cov += (lst_obj1[i] - mean1) * (lst_obj2[i] - mean2)
        if sample == False:
            cov = cov / n
            # Accounts for sample by dividing by one less than number of elemeints in the list
        else:
            cov = cov / (n - 1)
            return cov
    else:
        print("List lengths are not equal")
        print("List1:", n1)
        print("List2:", n2)
        
def correlation(lst_obj1, lst_obj2):
    # corr(x,y) = cov(x,y) / (SD(x) * SD(y))
    cov = covariance(lst_obj1, list_obj2)
    SD1 = SD(lst_obj1)
    SD2 = SD(lst_obj2)
    corr = cov / (SD1 * SD2)
    return corr

def skewness(lst_obj, sample = False):
    mean_ = mean(lst_obj)
    SD_ = SD(lst_obj, sample)
    skew = 0
    n = len(lst_obj)
    for val in lst_obj:
        skew += (val - mean) ** 3
    skew = skew / (n * SD_ **3) if not sample else n * skew / ((n - 1)*(n - 2) * SD_ ** 3)
    return skew

def kurtosis(lst_obj, sample = False):
    mean_ = mean(lst_obj)
    kurt = 0
    SD_ = SD(lst_obj, sample)
    n = len(lst_obj)
    for val in lst_obj:
        kurt += (val - mean_) ** 4
    kurt = kurt / (n * SD_ ** 4) if sample == False else n * (n + 1) * kurt / \
    ((n - 1) * (n - 2) * (n - 3) * (SD_ ** 4)) - (3 * (n-1) ** 2) / ((n - 2) * (n - 3))
    return kurt
```

```python
lst = [1, 2, 3, 4, 5, 8, 4]
print("Total:", total(lst))
print("Mean:", mean(lst))
print("Median:", median(lst))
print("Mode:", mode(lst))
print("Variance:", variance(lst))
print("Standard Deviation:", SD(lst))
```
